```python
import re

df['cleaned_answer'] = df['answer'].apply(
    lambda x: re.search(r'dasar hukum:(.*?)(?:Referensi|Informasi|$)', x, re.IGNORECASE | re.S).group(1).strip() if re.search(r'dasar hukum:(.*?)(?:Referensi|Informasi|$)', x, re.IGNORECASE | re.S) else 'Null'
)

df
```

```python
import re

def split_to_columns(text):
    # Menghapus tanda kurung siku dan tanda kutip, lalu pisahkan berdasarkan titik koma
    cleaned_text = re.sub(r"[\'\[\]]", "", text)  # Hilangkan tanda kurung dan kutip
    labels = [label.strip() for label in cleaned_text.split(";") if label.strip()]  # Pisahkan label
    return labels

# Terapkan fungsi ke kolom dasar_hukum
df['dasar_hukum_split'] = df['dasar_hukum'].apply(split_to_columns)

# Melihat hasil split
print(df['dasar_hukum_split'])
```
