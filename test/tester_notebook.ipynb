```python
import pandas as pd
import re
```

```python
df = pd.read_csv('../data/hukum_online_tech.csv')
df.head()
```

```python
df.info()
```

```python
df.duplicated().sum()
```

```python
df['cleaned_answer'] = df['answer'].apply(
    lambda x: re.search(r'dasar hukum:(.*?)(?:Referensi|$)', x, re.IGNORECASE | re.S).group(1).strip() if re.search(r'dasar hukum:(.*?)(?:Referensi|$)', x, re.IGNORECASE | re.S) else 'Null'
)

df
```

```python
null_data = df['cleaned_answer'] == 'Null'
null_data.sum()
```

```python
df = df[df['cleaned_answer'] != 'Null']
df
```

```python
df.loc[:, 'cleaned_answer'] = df['cleaned_answer'].str.replace(r'\d+\.\s*', ';', regex=True)
df.loc[:, 'cleaned_answer'] = df['cleaned_answer'].str.lstrip(';')
df
```

```python
null_data = df['cleaned_answer'] == 'Null'
null_data.sum()
```

```python
def clean_text(text):
    if ' No. ' not in text:
        text = text.split('.')[0] + '.'
        
    clean_text = text.split(';')
    return clean_text

df.loc[:, 'cleaned_answer'] = df['cleaned_answer'].apply(clean_text)

df
```

```python
df = df.drop(columns='answer')
df
```

```python
df_new = df.copy()
df_new
```

```python
pasal_set = set()
for pasal_list in df_new['cleaned_answer']:
    for pasal in pasal_list:
        if pasal:
            pasal_set.add(pasal)
            
pasal_list = list(pasal_set)
pasal_list.sort()
pasal_list
```

```python
for i, pasal in enumerate(pasal_list, start=1):
    df_new[f'{i}'] = df_new['cleaned_answer'].apply(lambda x: 1 if pasal in x else 0)
```
